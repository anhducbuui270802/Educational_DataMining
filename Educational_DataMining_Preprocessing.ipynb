#Bảng 4. xeploaiav.xlsx

##Xử lý chung

```python
# Đọc dữ liệu
df4 = pd.read_excel("Education_dataset_V2/04.xeploaiav.xlsx")
df4.head()
```

```python
# In ra các cột của bảng
df4.columns
```

```python
# Bỏ khoảng trắng ở tên cột
df4.columns = df4.columns.str.strip()
df4.columns
```

```python
df4.info()
```

```python
# Thống kê giá trị NaN ở mỗi cột
df4.isna().sum()
```

```python
#Xoá các dòng có giá trị NaN
df4 = df4.dropna()
df4.isna().sum()
```

##listening and reading

```python
df4['listening'] = df4['listening'].replace(' NULL', '0')
df4['reading'] = df4['reading'].replace(' NULL', '0')
```

```python
df4['listening'] = df4['listening'].astype(float)
df4['reading'] = df4['reading'].astype(float)
df4['total'] = df4['total'].astype(float)
```

```python
#thay thế các giá trị bị thiếu bằng total/2
df4['total'] = df4['total'].replace(0, np.nan)
df4['listening'] = df4['listening'].replace(0, np.nan)
df4['listening'] = df4['listening'].fillna(round(df4['total'].floordiv(2)))
```

```python
df4['reading'] = df4['reading'].replace(0, np.nan)
df4['reading'] = df4['reading'].fillna(df4['total'] - df4['listening'])
```

```python
df4
```

```python
df4 = df4.dropna()
```

##mamh

```python
df4['mamh'] = df4['mamh'].str.strip()
df4['mamh'].unique()
```

```python
#format lại mã lớp tiếng anh
df4['mamh'] = df4['mamh'].replace('EN001','ENG01')
df4['mamh'] = df4['mamh'].replace('EN002','ENG02')
df4['mamh'] = df4['mamh'].replace('AVSC','AVSC1')
df4['mamh'] = df4['mamh'].replace('','NULL')
```

```python
df4['mamh'] = df4['mamh'].where(df4['mamh'] != 'NULL', 
                              pd.cut(df4['total'], [0, 26, 40, 60, 70, 80, 90, float('inf')], 
                                     labels=['AVSC1', 'AVSC2', 'ENG01', 'ENG02', 'ENG03', 'ENG04', 'ENG05']))
```

```python
df4['mamh'].unique()
```

```python
df4[df4["ghichu"]==" Qui đổi TOEIC"]
```

##Lưu dữ liệu đã làm sạch

```python
df4.to_csv('data_clean/04.xeploaiav.csv', index=False)
```

#Bang 05.ThiSinh.xlsx

##Xử lý chung

```python
# Đọc dữ liệu
df5 = pd.read_excel("Education_dataset_V2/05.ThiSinh.xlsx")
df5.head()
```

```python
# Bỏ khoảng trắng ở tên cột
df5.columns = df5.columns.str.strip()
df5.columns
```

```python
df5.drop(columns=["lop12_matruong", "TEN_TRUONG"], inplace=True)
```

```python
df5["khuvuc"] = df5["lop12_matinh"].apply(lambda x: 1 if x==2 and not np.isnan(x)  
                                          else 0 if not np.isnan(x) 
                                          else np.nan)
```

```python
df5
```

```python
df5.info()
```

```python
df5.isna().sum()
```

###Xử lí khu vực (có ở TPHCM hay không)

```python
df5["dien_tt"].unique()
```

```python
df5.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

```python
# diện trúng tuyển là 30A thì khu vực là 0: ngoài TPHCM
df5["khuvuc"][df5["dien_tt"]=="30A"] = df5["khuvuc"][df5["dien_tt"]=="30A"].apply(lambda x: 0)
```

```python
# diện trúng tuyển là CCQT thì khu vực là 0: ngoài TPHCM
df5["khuvuc"][df5["dien_tt"]=="CCQT"] = df5["khuvuc"][df5["dien_tt"]=="CCQT"].apply(lambda x: 0)
```

```python
# diện trúng tuyển là CỬ TUYỂN thì khu vực là 0: ngoài TPHCM, dân tộc thiểu số
df5["khuvuc"][df5["dien_tt"]=="CUTUYEN"] = df5["khuvuc"][df5["dien_tt"]=="CUTUYEN"].apply(lambda x: 0)
```

```python
# diện trúng tuyển là THPT
mode_THPT = df5["khuvuc"][df5["dien_tt"]=="THPT"].mode()[0]
df5["khuvuc"][df5["dien_tt"]=="THPT"] = df5["khuvuc"][df5["dien_tt"]=="THPT"].fillna(mode_THPT)
```

```python
# diện trúng tuyển là TT-Bộ thì khu vực là 0: ngoài TPHCM, dân tộc thiểu số
df5["khuvuc"][df5["dien_tt"]=="TT-Bộ"] = df5["khuvuc"][df5["dien_tt"]=="TT-Bộ"].apply(lambda x: 0)
```

```python
df5["khuvuc"][df5["dien_tt"]=="ƯT-Bộ"] = df5["khuvuc"][df5["dien_tt"]=="ƯT-Bộ"].apply(lambda x: 0)
```

```python
# diện trúng tuyển là ƯT-ĐHQG
mode = df5["khuvuc"][df5["dien_tt"]=="ƯT-ĐHQG"].mode()[0]
df5["khuvuc"][df5["dien_tt"]=="ƯT-ĐHQG"] = df5["khuvuc"][df5["dien_tt"]=="ƯT-ĐHQG"].fillna(mode)
```

### Xử lí cột điểm, xếp loại

```python
df5.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

```python
df5["xl_tt"] = np.nan
df5
```

```python
# diện 30A
df5["xl_tt"][df5["dien_tt"]=="30A"] = df5["xl_tt"][df5["dien_tt"]=="30A"].apply(lambda x: 1)
```

```python
# diện CCQT
df5["xl_tt"][df5["dien_tt"]=="CCQT"] = df5["xl_tt"][df5["dien_tt"]=="CCQT"].apply(lambda x: 4)
```

```python
# diện CUTUYEN
df5["xl_tt"][df5["dien_tt"]=="CUTUYEN"] = df5["xl_tt"][df5["dien_tt"]=="CUTUYEN"].apply(lambda x: 1)
```

```python
# diện TT-Bộ
df5["xl_tt"][df5["dien_tt"]=="TT-Bộ"] = df5["xl_tt"][df5["dien_tt"]=="TT-Bộ"].apply(lambda x: 4)
```

```python
# diện ƯT-Bộ
df5["xl_tt"][df5["dien_tt"]=="ƯT-Bộ"] = df5["xl_tt"][df5["dien_tt"]=="ƯT-Bộ"].apply(lambda x: 4)
```

```python
#Kiểm tra trùng lặp
df5.duplicated().sum()
```

```python
df5.to_csv('data_clean/05.ThiSinh.csv', index=False)
```
