# Xử lý cột học kỳ

```python
df['hocky_sx'] = np.nan
```

```python
df_min_nam_hoc = df.groupby('mssv')['namhoc'].transform('min')
```

```python
df['hocky_sx'] = df.apply(lambda row: row['hocky'] + (row['namhoc'] - df_min_nam_hoc[row.name]) * 2, axis=1)
```

```python
df
```

```python
df[df["hocky_sx"] > 6] 
```

# Xử lý cột sotchk

```python
count_by_sotchk = df.groupby("sotchk").size()
print(count_by_sotchk)
```

* Những giá trị tín chỉ > 30 là không hợp lệ 
* Đưa chúng về = giá trị tín chỉ nhiều nhất (19)

```python
df.loc[df['sotchk'] > 30, 'sotchk'] = 19
```

```python
count_by_sotchk = df.groupby("sotchk").size()
print(count_by_sotchk)
```
