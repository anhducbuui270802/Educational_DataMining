```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

# Xử lí chung

```python
# Đọc dữ liệu
df = pd.read_excel("../Education_dataset_V2/04.xeploaiav.xlsx")
df.head()
```

```python
# In ra các cột của bảng
df.columns
```

```python
# Bỏ khoảng trắng ở tên cột
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# Kiểm tra giá trị thiếu
nan_rows = df[df.isna().any(axis=1)]
print(nan_rows)
```

```python
df.dropna(inplace=True)
```

# Xử lí reading và listening

```python
df['listening'].unique()
```

```python
df['reading'].unique()
```

```python
df['total'].unique()
```

```python
df['listening'] = df['listening'].replace(' NULL', '0')
df['reading'] = df['reading'].replace(' NULL', '0')
```

```python
# Chuyển kiểu dữ liệu
df['listening'] = df['listening'].astype(float)
df['reading'] = df['reading'].astype(float)
df['total'] = df['total'].astype(float)
```

```python
#thay thế các giá trị bị thiếu bằng total/2
df['total'] = df['total'].replace(0, np.nan)
df['listening'] = df['listening'].replace(0, np.nan)
df['listening'] = df['listening'].fillna(round(df['total'].floordiv(2)))
```

```python
df['reading'] = df['reading'].replace(0, np.nan)
df['reading'] = df['reading'].fillna(df['total'] - df['listening'])
```

```python
df.isna().sum()
```

```python
df['total'] = df['total'].replace(np.nan, 0)
df['listening'] = df['listening'].replace(np.nan, 0)
df['reading'] = df['reading'].replace(np.nan, 0)
```

```python
df
```

# Xử lí cột ghi chú

```python
df["ghichu"].unique()
```

```python
df["ghichu"] = df["ghichu"].str.strip()
df["ghichu"].unique()
```

```python
df["ghichu"] = df["ghichu"].replace("NULL);", "NULL")
df["ghichu"] = df["ghichu"].replace(["Qui đổi TOEIC);", "'Qui đổi TOEIC'"], "Qui đổi TOEIC")
df["ghichu"].unique()
```

# Xử lí xếp loại anh văn

```python
df["total"].unique()
```

```python
df["total"][df["ghichu"]=="Qui đổi TOEIC"].unique()
```

```python
df["total"][df["ghichu"]=="NULL"].unique()
```

```python
df["xl_av"] = np.nan
```

## Xử lí ghichu Qui đổi TOEIC

```python
def convert_TOEIC_to_xeploai(x):
  if x < 300:
    return 0
  elif x >= 300 and x <= 345:
    return 1
  elif x >= 350 and x <= 395:
    return 2
  elif x >= 400 and x <= 445:
    return 3
  elif x >= 500 and x <= 555:
    return 4
  else:
    return 5
df["xl_av"][df["ghichu"]=="Qui đổi TOEIC"] = df["total"][df["ghichu"]=="Qui đổi TOEIC"].apply(lambda x: convert_TOEIC_to_xeploai(x))
```

## Xử lí ghichu bằng NULL nhưng có điểm total > 120 
## (số điểm anh văn đầu vào tối đa là 120, vượt quá 120 => qui đổi TOEIC)

```python
df["xl_av"][(df["ghichu"] == "NULL") & (df['total'] > 120)] = df["total"][(df["ghichu"] == "NULL") & (df['total'] > 120)].apply(lambda x: convert_TOEIC_to_xeploai(x))
```

```python
df[(df["ghichu"] == "NULL") & (df['total'] > 120)]
```

## Xử lí ghichu bằng NULL

```python
def convert_avdv_to_xeploai(x):
  if x < 40:
    return 0
  elif x >= 40 and x <= 60:
    return 1
  elif x >= 60 and x <= 70:
    return 2
  elif x >= 70 and x <= 80:
    return 3
  elif x >= 80 and x <= 90:
    return 4
  else:
    return 5
df["xl_av"][(df["ghichu"] == "NULL") & (df['total'] <= 120)] = df["total"][(df["ghichu"] == "NULL") & (df['total'] <= 120)].apply(lambda x: convert_avdv_to_xeploai(x))
```

```python
df[(df["ghichu"] == "NULL") & (df['total'] <= 120)]
```

```python
df.drop(columns=["id", "listening", "reading", "mamh", "ghichu"], inplace=True)
```
