# Xử lí xếp loại đrl

```python
df["xl_drl"] = np.nan
```

```python
def convert_DRL_to_xeploai(x):
  if x < 35:
    return 0
  elif x >= 35 and x < 50:
    return 1
  elif x >= 50 and x < 65:
    return 2
  elif x >= 65 and x < 80:
    return 3
  elif x >= 80 and x < 90:
    return 4
  elif x >= 90:
    return 5
```

```python
df["xl_drl"] = df["drl"].apply(lambda x: convert_DRL_to_xeploai(x))
df
```
