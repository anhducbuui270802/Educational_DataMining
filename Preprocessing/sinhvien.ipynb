```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

# Xử lí chung

```python
# Đọc dữ liệu
df = pd.read_excel("../Education_dataset_V2/01.sinhvien.xlsx")
df.head()
```

```python
# In ra các cột của bảng
df.columns
```

```python
# Bỏ các cột bắt đầu bằng kí tự "_"
df = df.drop(df.columns[df.columns.str.startswith('_')], axis=1)
df.columns
```

```python
# Bỏ khoảng trắng ở tên cột
df.columns = df.columns.str.strip()
df.columns
```

```python
# Thống kê giá trị NaN ở mỗi cột
df.isna().sum()
```

```python
df.shape
```

```python
# Cột Column1 có nhiều giá trị NaN
print("Số lượng giá trị khác NaN trong cột NaN:", 
      len(df['Column1'][df['Column1'].notna()].unique()))
df['Column1'][df['Column1'].notna()].unique()
```

```python
# Bỏ cột Column1 vì có nhiều giá trị NaN
df.drop(columns = "Column1", inplace=True)
df.head()
```

```python
# Thống kê giá trị NaN ở mỗi cột
df[df.isna().any(axis=1)]
```

```python
# Loại bỏ những dòng có mssv là NaN
df.dropna(subset=["mssv"], inplace=True)
```

# Xử lí nơi sinh

```python
df["noisinh"].unique()
```

```python
# Loại bỏ kí tự ' ở 2 đầu giá trị (thay thế ' bằng kí tự rỗng)
df['noisinh'] = df['noisinh'].str.replace("'", '', regex=True)
df["noisinh"].unique()
```

```python
# Định dạng in thường
df['noisinh'] = df['noisinh'].str.lower()
df["noisinh"].unique()
```

```python
# Loại bỏ kí tự tp. ở đầu giá trị (thay thế tp. bằng kí tự rỗng)
df['noisinh'] = df['noisinh'].str.replace("tp.", '', regex=True)
df['noisinh'] = df['noisinh'].str.replace("thành phố.", '', regex=True)
df['noisinh'] = df['noisinh'].str.replace("tỉnh", '', regex=True)
df["noisinh"].unique()
```

```python
# Loại bỏ khoảng trắng ở 2 đầu giá trị
df["noisinh"] = df["noisinh"].str.strip()
df["noisinh"].unique()
```

# Xử lí lopsh
