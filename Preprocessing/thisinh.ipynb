# Xử lí chung

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Đọc dữ liệu
df = pd.read_excel("../Education_dataset_V2/05.ThiSinh.xlsx")
df.head()
```

```python
# Bỏ khoảng trắng ở tên cột
df.columns = df.columns.str.strip()
df.columns
```

```python
df.drop(columns=["lop12_matruong", "TEN_TRUONG"], inplace=True)
```

```python
# Thêm cột khu vực: 1 là thuộc TPHCM, 0 là khu vực khác
df["khuvuc"] = df["lop12_matinh"].apply(lambda x: 1 if x==2 and not np.isnan(x)  
                                          else 0 if not np.isnan(x) 
                                          else np.nan)
df.head()
```

```python
df["dien_tt"].unique()
```

```python
# Kiểm tra giá trị thiếu theo diện trúng tuyển
df.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

# Xử lí khu vực trúng tuyển

```python
# dien_tt = 30A thì khuvuc = 0, dân tộc thiểu số
df["khuvuc"][df["dien_tt"]=="30A"] = df["khuvuc"][df["dien_tt"]=="30A"].apply(lambda x: 0)
```

```python
# dien_tt = CCQT thì khuvuc = 0
df["khuvuc"][df["dien_tt"]=="CCQT"] = df["khuvuc"][df["dien_tt"]=="CCQT"].apply(lambda x: 0)
```

```python
# dien_tt = CUTUYEN thì khuvuc = 0: ngoài TPHCM, dân tộc thiểu số
df["khuvuc"][df["dien_tt"]=="CUTUYEN"] = df["khuvuc"][df["dien_tt"]=="CUTUYEN"].apply(lambda x: 0)
```

```python
# dien_tt = THPT thì khuvuc = giá trị khuvuc xuất hiện nhiều nhất trong diện THPT
mode = df["khuvuc"][df["dien_tt"]=="THPT"].mode()[0]
df["khuvuc"][df["dien_tt"]=="THPT"] = df["khuvuc"][df["dien_tt"]=="THPT"].fillna(mode)
```

```python
# dien_tt = TT-Bộ thì khuvuc = 0
df["khuvuc"][df["dien_tt"]=="TT-Bộ"] = df["khuvuc"][df["dien_tt"]=="TT-Bộ"].apply(lambda x: 0)
```

```python
# dien_tt = ƯT-Bộ thì khuvuc = 0
df["khuvuc"][df["dien_tt"]=="ƯT-Bộ"] = df["khuvuc"][df["dien_tt"]=="ƯT-Bộ"].apply(lambda x: 0)
```

```python
# diện trúng tuyển là ƯT-ĐHQG
mode = df["khuvuc"][df["dien_tt"]=="ƯT-ĐHQG"].mode()[0]
df["khuvuc"][df["dien_tt"]=="ƯT-ĐHQG"] = df["khuvuc"][df["dien_tt"]=="ƯT-ĐHQG"].fillna(mode)
```

```python
# Kiểm tra giá trị thiếu theo diện trúng tuyển
df.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

# Xử lí điểm, xếp loại

```python
# Kiểm tra giá trị thiếu theo diện trúng tuyển
df.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

```python
df["xl_tt"] = np.nan
```

```python
# diện 30A
df["xl_tt"][df["dien_tt"]=="30A"] = df["xl_tt"][df["dien_tt"]=="30A"].apply(lambda x: 1)
```

```python
# diện CCQT
df["xl_tt"][df["dien_tt"]=="CCQT"] = df["xl_tt"][df["dien_tt"]=="CCQT"].apply(lambda x: 4)
```

```python
# diện CUTUYEN
df["xl_tt"][df["dien_tt"]=="CUTUYEN"] = df["xl_tt"][df["dien_tt"]=="CUTUYEN"].apply(lambda x: 1)
```

```python
# diện TT-Bộ
df["xl_tt"][df["dien_tt"]=="TT-Bộ"] = df["xl_tt"][df["dien_tt"]=="TT-Bộ"].apply(lambda x: 4)
```

```python
# diện ƯT-Bộ
df["xl_tt"][df["dien_tt"]=="ƯT-Bộ"] = df["xl_tt"][df["dien_tt"]=="ƯT-Bộ"].apply(lambda x: 4)
```

```python
# diện ƯT-ĐHQG
df["xl_tt"][df["dien_tt"]=="ƯT-ĐHQG"] = df["xl_tt"][df["dien_tt"]=="ƯT-ĐHQG"].apply(lambda x: 4)
```

```python
# diện THPT
# xử lí giá trị thiếu
mode = df["diem_tt"][df["dien_tt"]=="THPT"].mean()
df["diem_tt"][df["dien_tt"]=="THPT"] = df["diem_tt"][df["dien_tt"]=="THPT"].fillna(mode)
```

```python
# xử lí xếp loại
df["xl_tt"][df["dien_tt"]=="THPT"] = df["diem_tt"][df["dien_tt"]=="THPT"].apply(lambda x: 4 if x >= 28 else 3 if x >= 25 else 2 if x >= 22 else 1 if x >= 20 else 0)
```

```python
# diện ĐGNL
# xử lí xếp loại
def convert_to_xep_loai(diem):
    if diem >= 1000:
      return 4
    elif diem >= 900:
      return 3
    elif diem >= 750:
      return 2
    elif diem >= 600:
      return 1
    else:
      return 0
df["xl_tt"][df["dien_tt"]=="ĐGNL"] = df["diem_tt"][df["dien_tt"]=="ĐGNL"].apply(lambda x: convert_to_xep_loai(x))
```

```python
# Kiểm tra giá trị thiếu theo diện trúng tuyển
df.groupby("dien_tt").apply(lambda x: x.isna().sum())
```
