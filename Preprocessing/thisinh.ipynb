```python
# dien_tt = 30A thì khuvuc = 5, khu vực Tây Nguyên
merged_df.loc[merged_df['dien_tt'] == '30A', 'khuvuc_x'] = merged_df.loc[merged_df['dien_tt'] == '30A', 'khuvuc_x'].fillna(5)
```

```python
# dien_tt = CUTUYEN thì khuvuc = 5, khu vực Tây Nguyên
merged_df.loc[merged_df['dien_tt'] == 'CUTUYEN', 'khuvuc_x'] = merged_df.loc[merged_df['dien_tt'] == 'CUTUYEN', 'khuvuc_x'].fillna(5)
```

```python
# dien_tt = CCQT thì khuvuc = 5, khu vực Tây Nguyên
merged_df.loc[merged_df['dien_tt'] == 'CCQT', 'khuvuc_x'] = merged_df.loc[merged_df['dien_tt'] == 'CCQT', 'khuvuc_x'].fillna(5)
```

```python
# dien_tt = ƯT-Bộ thì khuvuc = 5, khu vực Tây Nguyên
merged_df.loc[merged_df['dien_tt'] == 'ƯT-Bộ', 'khuvuc_x'] = merged_df.loc[merged_df['dien_tt'] == 'ƯT-Bộ', 'khuvuc_x'].fillna(5)
```

```python
# dien_tt = TT-Bộ thì khuvuc = 5, khu vực Tây Nguyên
merged_df.loc[merged_df['dien_tt'] == 'TT-Bộ', 'khuvuc_x'] = merged_df.loc[merged_df['dien_tt'] == 'TT-Bộ', 'khuvuc_x'].fillna(5)
```

```python
# diện trúng tuyển là ƯT-ĐHQG, khuvuc được gán bởi giá trị xuất hiện nhiều nhất
mode = merged_df["khuvuc_x"][merged_df["dien_tt"]=="ƯT-ĐHQG"].mode()[0]
merged_df["khuvuc_x"][merged_df["dien_tt"]=="ƯT-ĐHQG"] = merged_df["khuvuc_x"][merged_df["dien_tt"]=="ƯT-ĐHQG"].fillna(mode)
```

```python
# diện trúng tuyển là THPT, khuvuc được gán bởi giá trị xuất hiện nhiều nhất
mode = merged_df["khuvuc_x"][merged_df["dien_tt"]=="THPT"].mode()[0]
merged_df["khuvuc_x"][merged_df["dien_tt"]=="THPT"] = merged_df["khuvuc_x"][merged_df["dien_tt"]=="THPT"].fillna(mode)
```

## Xử lý xl_tt

```python
# diện 30A
merged_df["xl_tt"][merged_df["dien_tt"]=="30A"] = merged_df["xl_tt"][merged_df["dien_tt"]=="30A"].apply(lambda x: 1)
```

```python
# diện CCQT
merged_df["xl_tt"][merged_df["dien_tt"]=="CCQT"] = merged_df["xl_tt"][merged_df["dien_tt"]=="CCQT"].apply(lambda x: 4)
```

```python
# diện CUTUYEN
merged_df["xl_tt"][merged_df["dien_tt"]=="CUTUYEN"] = merged_df["xl_tt"][merged_df["dien_tt"]=="CUTUYEN"].apply(lambda x: 1)
```

```python
# diện TT-Bộ
merged_df["xl_tt"][merged_df["dien_tt"]=="TT-Bộ"] = merged_df["xl_tt"][merged_df["dien_tt"]=="TT-Bộ"].apply(lambda x: 4)
```

```python
# diện ƯT-Bộ
merged_df["xl_tt"][merged_df["dien_tt"]=="ƯT-Bộ"] = merged_df["xl_tt"][merged_df["dien_tt"]=="ƯT-Bộ"].apply(lambda x: 4)
```

```python
# diện ƯT-ĐHQG
merged_df["xl_tt"][merged_df["dien_tt"]=="ƯT-ĐHQG"] = merged_df["xl_tt"][merged_df["dien_tt"]=="ƯT-ĐHQG"].apply(lambda x: 4)
```

```python
# diện THPT
# xử lí giá trị thiếu
mode = merged_df["diem_tt"][merged_df["dien_tt"]=="THPT"].mean()
merged_df["diem_tt"][merged_df["dien_tt"]=="THPT"] = merged_df["diem_tt"][merged_df["dien_tt"]=="THPT"].fillna(mode)
```

```python
# diện ĐGNL
# xử lí xếp loại
def convert_to_xep_loai(diem):
    if diem >= 1000:
      return 4
    elif diem >= 900:
      return 3
    elif diem >= 750:
      return 2
    elif diem >= 600:
      return 1
    else:
      return 0
merged_df["xl_tt"][merged_df["dien_tt"]=="ĐGNL"] = merged_df["diem_tt"][merged_df["dien_tt"]=="ĐGNL"].apply(lambda x: convert_to_xep_loai(x))
```

```python
# Kiểm tra giá trị thiếu theo diện trúng tuyển
merged_df.groupby("dien_tt").apply(lambda x: x.isna().sum())
```

```python
merged_df_THPT = merged_df[merged_df["dien_tt"]=="THPT"]
print(merged_df_THPT.shape[0])
merged_df_THPT.isna().sum()
```

```python
# xử lí xếp loại điểm THPT
# Vì mỗi khoá có phổ điểm khác nhau nên sẽ phân loại thành 5 mức theo từng khoá học
# Tạo hàm áp dụng cho phương thức apply()
def binning(x):
    bins = np.linspace(min(x), max(x), 6)
    labels = [0,1,2,3,4]
    return pd.cut(x, bins=bins, labels=labels, include_lowest=True)
merged_df_THPT['xl_tt'] = merged_df_THPT.groupby('khoahoc')['diem_tt'].apply(binning)
```

```python
merged_df = pd.merge(merged_df, merged_df_THPT[['mssv', 'xl_tt']], on='mssv', how='left')
```

```python
merged_df['xl_tt_x'] = merged_df['xl_tt_x'].fillna(merged_df['xl_tt_y'])
```

```python
# Xóa cột 'xl_tt_y' sau khi sử dụng
merged_df = merged_df.drop('xlav_y', axis=1)

# # Đổi tên cột 'xl_tt_x' thành 'xl_tt'
merged_df = merged_df.rename(columns={'xl_tt_x': 'xl_tt'})
```
