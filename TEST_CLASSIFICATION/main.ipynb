# Điểm av đầu vào

```python
df_av = pd.read_csv("../data_clean/04.xeploaiav.csv")
df_av.head()
```

```python
df_av.drop(columns=["total"], inplace=True)
```

# m2 = m1 & Điểm av

```python
merge_df2 = pd.merge(merge_df1, df_av, on="mssv", how="left")
merge_df2.head()
```

```python
# Thống kê giá trị thiếu
merge_df2.isna().sum()
```

```python
# Fillna "xl_av" bằng giá trị trung bình
mode_xl_av = int(round(merge_df2["xl_av"].mean()))
merge_df2["xl_av"] = merge_df2["xl_av"].fillna(mode_xl_av)
```

```python
merge_df2.shape
```

# Điểm trung bình học kỳ

```python
df_dtbhk = pd.read_csv("../data_clean/sinhvien_dtb_hocky.csv")
df_dtbhk.head()
```

```python
df_dtbhk.drop(columns=["hocky_sx"], inplace=True)
```

# Điểm rèn luyện

```python
df_drl = pd.read_csv("../data_clean/07.diemrl.csv")
df_drl.head()
```

```python
df_drl.drop(columns=["id"], inplace=True)
df_drl.head()
```

# m3 = dtb & drl

```python
merge_df3 = pd.merge(df_dtbhk, df_drl, on=["mssv", "namhoc", "hocky"], how="left")
merge_df3
```

```python
# Bỏ dữ liệu học kỳ hè
merge_df3 = merge_df3.drop(merge_df3[merge_df3["hocky"] == 3].index)
merge_df3.shape
```

```python
# Sắp xếp các bản ghi theo mỗi sinh viên và năm học, học kỳ
merge_df3 = merge_df3.sort_values(by=['mssv', 'namhoc', 'hocky'])
merge_df3.head()
```

# m4 = m2 & m3

```python
merge_df4 = pd.merge(merge_df2, merge_df3, on="mssv", how="inner")
merge_df4
```

```python
merge_df4.isna().sum()
```

```python
# Điền giá trị thiếu
# merge_df4["xl_drl"] = merge_df4.groupby("mssv")["xl_drl"].transform(lambda x: x.fillna(x.mean() if x.notna().any() else 0))
merge_df4["drl"] = merge_df4.groupby("mssv")["drl"].transform(lambda x: x.fillna(x.mean() if x.notna().any() else 0))
merge_df4
```

```python
# Tạo cột số thứ tự của học kỳ
merge_df4["hocky_sx"] = merge_df4.groupby("mssv").cumcount()+1
merge_df4.head()
```

```python
# dataframe chính
df = merge_df4.copy()
```

```python
# Điểm trung bình của các học kì trước
sum_dtbtl = df.groupby("mssv")["dtbhk"].cumsum().shift()
sum_sohk = df.groupby('mssv')['hocky_sx'].cumcount()
df["dtbtl"] = sum_dtbtl/sum_sohk
df["dtbtl"]
```

```python
# Điểm rèn luyện trung bình
sum_drltl = df.groupby("mssv")["drl"].cumsum().shift()
sum_sohk = df.groupby('mssv')['hocky_sx'].cumcount()
df["drltl"] = sum_drltl/sum_sohk
df["drltl"]
```

```python
# Bỏ các dòng học kỳ 1
df.drop(df[df["hocky_sx"]==1].index, inplace=True)
df.head()
```

```python
df.drop(columns="xl_drl", inplace=True)
```

```python
# Tạo danh sách các khoảng giá trị và xếp loại tương ứng
# bins = [0, 4, 5, 7, 8, 9, 10.1]
# labels = [0, 1, 2, 3, 4, 5]

bins = [0, 5, 7, 8, 10.1]
labels = [0, 1, 2, 3]

# Áp dụng pd.cut để xếp loại giá trị của cột "diemtb"
df["xl_dtbhk"] = pd.cut(df["dtbhk"], bins=bins, labels=labels, include_lowest=True, right=False)
df.drop(columns="dtbhk", inplace=True)
```

```python
df["xl_dtbhk"].value_counts()
```

```python
df.head()
```

```python
def convert_DRL_to_xeploai(x):
  if x < 35:
    return 0
  elif x >= 35 and x < 50:
    return 1
  elif x >= 50 and x < 65:
    return 2
  elif x >= 65 and x < 80:
    return 3
  elif x >= 80 and x < 90:
    return 4
  elif x >= 90:
    return 5
```

```python
df["drltl"] = df["drltl"].apply(lambda x: convert_DRL_to_xeploai(x))
```

# One hot

```python
# One hot giá trị hệ đào tạo
one_hot_df = pd.get_dummies(df['hedt'], prefix='hedt')

df = df.iloc[:, :2].join(one_hot_df.astype(int)).join(df.iloc[:, 2:])
df.drop(columns=["hedt"], inplace=True)
df.head()
```

```python
drop_columns = ["hocky", "drl"]
df.drop(columns=drop_columns, inplace=True)
```
