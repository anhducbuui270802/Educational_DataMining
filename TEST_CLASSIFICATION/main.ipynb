# Điểm trung bình học kỳ

```python
df_dtbhk = pd.read_csv("../data_clean/sinhvien_dtb_hocky.csv")
df_dtbhk.head()
```

# Điểm rèn luyện

```python
df_drl = pd.read_csv("../data_clean/07.diemrl.csv")
df_drl.head()
```

```python
df_drl.drop(columns=["id", "drl"], inplace=True)
df_drl.head()
```

# Merge3 dtb & drl

```python
merge_df3 = pd.merge(df_dtbhk, df_drl, on=["mssv", "namhoc", "hocky"], how="left")
merge_df3
```

```python
# Bỏ dữ liệu học kỳ hè
merge_df3 = merge_df3.drop(merge_df3[merge_df3["hocky"] == 3].index)
merge_df3.shape
```

```python
# Lọc những sinh viên có số học kỳ đã học >= 7
filtered_sv = merge_df3["mssv"][merge_df3["hocky_sx"]>=7].unique()
filtered_sv
```

```python
# Lọc các dòng theo điều kiện: sinh viên có số học kỳ đã học >= 7 và hocky_sx <=7
df_filtered = merge_df3[(merge_df3["mssv"].isin(filtered_sv)) 
                        & (merge_df3["hocky_sx"]<=7)]
df_filtered
```

## Pivot theo dtbhk

```python
# Tạo pivot table
pv_dtbhk = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="dtbhk")

# Chuyển pivot_table về dataframe với các cột theo học kỳ
pv_dtbhk = pv_dtbhk.reset_index()
pv_dtbhk.columns.name = None
pv_dtbhk.columns = ["mssv"] + ["dtbhk" + str(col) for col in pv_dtbhk.columns[1:]]
pv_dtbhk.head()
```

## Pivot theo drl

```python
# Tạo pivot table
pv_drl = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="xl_drl")

# Chuyển pivot_table về dataframe với các cột theo học kỳ
pv_drl = pv_drl.reset_index()
pv_drl.columns.name = None
pv_drl.columns = ["mssv"] + ["xl_drl" + str(col) for col in pv_drl.columns[1:]]
pv_drl.head()
```

```python
# merge dtbhk & drl
merge_pv = pd.merge(pv_dtbhk, pv_drl, on="mssv", how="outer")
merge_pv.head()
```

## Pivot theo sotchk

```python
# Tạo pivot table
pv_sotchk = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="sotchk")

# Chuyển pivot_table về dataframe với các cột theo học kỳ
pv_sotchk = pv_sotchk.reset_index()
pv_sotchk.columns.name = None
pv_sotchk.columns = ["mssv"] + ["sotchk" + str(col) for col in pv_sotchk.columns[1:]]
pv_sotchk.head()
```

```python
# merge dtbhk & drl & sotchk
merge_pv = pd.merge(merge_pv, pv_sotchk, on="mssv", how="outer")
merge_pv.head()
```

```python
# Thống kê giá trị thiếu
merge_pv.isna().sum()
```

```python
# merge_pv.dropna(inplace=True)
```

```python
# Fill nan
for col in merge_pv.columns[1:]:
    if col.startswith('dtbhk'):
        value = merge_pv[col].mean()
        merge_pv[col] = merge_pv[col].fillna(value)
    else:
        value = int(round(merge_pv[col].mean()))
        merge_pv[col] = merge_pv[col].fillna(value)
```

# Merge sinhvien, thisinh & diem

```python
result = pd.merge(merge_df2, merge_pv, on="mssv", how="inner")
result
```

```python
# Tạo danh sách các khoảng giá trị và xếp loại tương ứng
bins = [0, 4, 5, 7, 8, 9, 10.1]
labels = [0, 1, 2, 3, 4, 5]

# Áp dụng pd.cut để xếp loại giá trị của cột "diemtb"
for col in result.columns:
    if col.startswith('dtbhk'):
        result[col] = pd.cut(result[col], bins=bins, labels=labels, include_lowest=True, right=False)
```

```python
# Chuyển giá trị khoa về dạng mã hóa one hot
one_hot_df = pd.get_dummies(result['khoa'], prefix='khoa')

# Kết hợp DataFrame one-hot encoding với merge_df4
merge_df5 = result.iloc[:, :2].join(one_hot_df.astype(int)).join(result.iloc[:, 2:])
merge_df5.drop(columns=["khoa"], inplace=True)
merge_df5.head()
```

```python
# Chuyển giá trị hệ đào tạo về dạng one hot
one_hot_df = pd.get_dummies(merge_df5['hedt'], prefix='hedt')

# Kết hợp DataFrame one-hot encoding với merge_df5
merge_df5 = merge_df5.iloc[:, :2].join(one_hot_df.astype(int)).join(merge_df5.iloc[:, 2:])
merge_df5.drop(columns=["hedt"], inplace=True)
merge_df5.head()
```
