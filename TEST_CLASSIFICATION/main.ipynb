# dtbhk

```python
df_dtbhk = pd.read_csv("../data_clean/sinhvien_dtb_hocky.csv")
df_dtbhk
```

```python
# In các học kỳ
df_dtbhk["hocky"].value_counts()
```

```python
# Bỏ dữ liệu học kỳ hè
df_dtbhk = df_dtbhk.drop(df_dtbhk[df_dtbhk["hocky"] == 3].index)
```

```python
# Lọc những sinh viên có số học kỳ chính >= 7
df_filtered1 = df_dtbhk[df_dtbhk["hocky_sx"]>=7]
df_filtered1
```

```python
# Lọc các hàng trong danh sách ban đầu có mssv thuộc df_filtered1
df_filtered2 = df_dtbhk[df_dtbhk["mssv"].isin(df_filtered1["mssv"].unique())]
df_filtered2
```

```python
# Chỉ lọc từ học kỳ 1 đến 7 để huấn luyện
df_filtered = df_filtered2[df_filtered2["hocky_sx"]<=7]
df_filtered
```

```python
pv_dtbhk = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="dtbhk")
pv_dtbhk
```

```python
df_dtbhk_result = pv_dtbhk.reset_index()
df_dtbhk_result.columns.name = None
df_dtbhk_result.columns = ["mssv"] + ["dtb_hk" + str(col) for col in df_dtbhk_result.columns[1:]]
df_dtbhk_result
```

```python
df_dtbhk_result.isna().sum()
```

```python
# Tính giá trị trung bình từng cột
mean_dtb_hk2 = df_dtbhk_result['dtb_hk2'].mean()
mean_dtb_hk3 = df_dtbhk_result['dtb_hk3'].mean()
mean_dtb_hk4 = df_dtbhk_result['dtb_hk4'].mean()
mean_dtb_hk5 = df_dtbhk_result['dtb_hk5'].mean()
mean_dtb_hk6 = df_dtbhk_result['dtb_hk6'].mean()
mean_dtb_hk7 = df_dtbhk_result['dtb_hk7'].mean()

# Fillna cho từng cột bằng giá trị trung bình tương ứng
df_dtbhk_result['dtb_hk2'].fillna(mean_dtb_hk2, inplace=True)
df_dtbhk_result['dtb_hk3'].fillna(mean_dtb_hk3, inplace=True)
df_dtbhk_result['dtb_hk4'].fillna(mean_dtb_hk4, inplace=True)
df_dtbhk_result['dtb_hk5'].fillna(mean_dtb_hk5, inplace=True)
df_dtbhk_result['dtb_hk6'].fillna(mean_dtb_hk6, inplace=True)
df_dtbhk_result['dtb_hk7'].fillna(mean_dtb_hk7, inplace=True)
```

# merge dtb&drl

```python
merge_df2 = pd.merge(df_dtbhk, df_drl, on=["mssv", "namhoc", "hocky"], how="left")
merge_df2
```

```python
# Bỏ dữ liệu học kỳ hè
merge_df2 = merge_df2.drop(merge_df2[merge_df2["hocky"] == 3].index)
merge_df2.shape
```

```python
# Thống kê dữ liệu thiếu
merge_df2.isna().sum()
```

```python
# Lọc những sinh viên có số học kỳ đã học >= 7
filtered_sv = merge_df2["mssv"][merge_df2["hocky_sx"]>=7].unique()
filtered_sv
```

```python
df_filtered = merge_df2[(merge_df2["mssv"].isin(filtered_sv)) 
                        & (merge_df2["hocky_sx"]<=7)]
df_filtered
```

## pivot_dtbhk

```python
pv_dtbhk = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="dtbhk")
pv_dtbhk = pv_dtbhk.reset_index()
pv_dtbhk.columns.name = None
pv_dtbhk.columns = ["mssv"] + ["dtbhk" + str(col) for col in pv_dtbhk.columns[1:]]
pv_dtbhk
```

## pv_drl

```python
pv_drl = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="xl_drl")
pv_drl = pv_drl.reset_index()
pv_drl.columns.name = None
pv_drl.columns = ["mssv"] + ["xl_drl" + str(col) for col in pv_drl.columns[1:]]
pv_drl
```

```python
merge_pv = pd.merge(pv_dtbhk, pv_drl, on="mssv", how="outer")
merge_pv
```

## pv_sotchk

```python
pv_sotchk = df_filtered.pivot_table(index="mssv", columns="hocky_sx", values="sotchk")
pv_sotchk = pv_sotchk.reset_index()
pv_sotchk.columns.name = None
pv_sotchk.columns = ["mssv"] + ["sotchk" + str(col) for col in pv_sotchk.columns[1:]]
pv_sotchk
```

```python
merge_pv = pd.merge(merge_pv, pv_sotchk, on="mssv", how="outer")
merge_pv
```

```python
merge_pv.dropna(inplace=True)
merge_pv
```
