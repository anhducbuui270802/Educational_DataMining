# m2 = m1 & Điểm av

```python
nan_row = merge_df2[["mssv","xl_av"]][merge_df2['xl_av'].isna()]
nan_row
```

Thay thế các giá trị NaN trong cột phân loại bằng cách phân loại điểm học tập môn ENG đầu tiên mà sinh viên đó học

```python
# Đọc dữ liệu điểm các môn học từ file
df_diemThu = pd.read_excel("../Education_dataset_V2/diem_Thu.xlsx")
df_diemThu = df_diemThu[['mssv','mamh','diem_hp']]
df_diemThu
```

```python
#Tìm những sinh viên bị thiếu ở cột xếp loại anh văn ở bảng merge_df2
#Từ những sinh viên đó tìm trong bảng diemThu các môn mà sinh viên đó học
df_merged = pd.merge(nan_row, df_diemThu, on='mssv')
df_merged
```

```python
#Lọc ra những môn tiếng anh mà sinh viên đó học, (chỉ lấy mã môn đầu tiên)
df_filtered = df_merged[df_merged['mamh'].str.startswith(('ENG', 'AV'))]
df_filtered = df_filtered.drop_duplicates(subset=['mssv'], keep='first')
replace = {'ENGL1113': 'ENG01', 'ENGA2': 'ENG02', 'ENGA1': 'ENG01'}
df_filtered['mamh'] = df_filtered['mamh'].replace(replace)
df_filtered
```

```python
# Hàm dùng để phân loại trình độ tiếng anh của sinh viên
def update_xl(row):
    if row['mamh'] == 'ENG01' and row['diem_hp'] < 5:
        return 0
    elif row['mamh'] == 'ENG01' and row['diem_hp'] >= 5:
        return 1
    elif row['mamh'] == 'ENG02' and row['diem_hp'] < 5:
        return 1
    elif row['mamh'] == 'ENG02' and row['diem_hp'] >= 5:
        return 2
    elif row['mamh'] == 'ENG03' and row['diem_hp'] < 5:
        return 2
    elif row['mamh'] == 'ENG03' and row['diem_hp'] >= 5:
        return 3
    elif row['mamh'] == 'ENG04' and row['diem_hp'] < 5:
        return 3
    elif row['mamh'] == 'ENG04' and row['diem_hp'] >= 5:
        return 4
    elif row['mamh'] == 'ENG05' and row['diem_hp'] < 5:
        return 4
    elif row['mamh'] == 'ENG05' and row['diem_hp'] >= 5:
        return 5
    else:
        return 0
```

```python
df_filtered['xl_av'] =df_filtered.apply(lambda row: update_xl(row), axis=1)
```

```python
merge_df2 = pd.merge(merge_df2, df_filtered[['mssv','xl_av']], on="mssv", how="left")

merge_df2['xl_av_x'] = merge_df2['xl_av_x'].fillna(merge_df2['xl_av_y'])
# Xóa cột 'xlav_y' sau khi sử dụng
merge_df2 = merge_df2.drop('xl_av_y', axis=1)

# Đổi tên cột 'xlav_x' thành 'xlav'
merge_df2 = merge_df2.rename(columns={'xl_av_x': 'xl_av'})
```

```python
# Fillna "xl_av" bằng giá trị trung bình
mode_xl_av = int(round(merge_df2["xl_av"].mean()))
merge_df2["xl_av"] = merge_df2["xl_av"].fillna(mode_xl_av)
```

# Điểm trung bình học kỳ

```python
df_dtbhk = pd.read_csv("../data_clean/sinhvien_dtb_hocky.csv")
df_dtbhk.head()
```

```python
df_dtbhk.drop(columns=["hocky_sx"], inplace=True)
```

# Điểm rèn luyện

```python
df_drl = pd.read_csv("../data_clean/07.diemrl.csv")
df_drl.head()
```

```python
df_drl.drop(columns=["id", "xl_drl"], inplace=True)
df_drl.head()
```

# m3 = dtb & drl

```python
merge_df3 = pd.merge(df_dtbhk, df_drl, on=["mssv", "namhoc", "hocky"], how="left")
merge_df3
```

```python
# Bỏ dữ liệu học kỳ hè
merge_df3 = merge_df3.drop(merge_df3[merge_df3["hocky"] == 3].index)
merge_df3.shape
```

```python
# Sắp xếp các bản ghi theo mỗi sinh viên và năm học, học kỳ
merge_df3 = merge_df3.sort_values(by=['mssv', 'namhoc', 'hocky'])
merge_df3.head()
```

# m4 = m2 & m3

```python
merge_df4 = pd.merge(merge_df2, merge_df3, on="mssv", how="inner")
merge_df4
```

```python
merge_df4.isna().sum()
```

```python
# Điền giá trị thiếu ở cột điểm rèn luyện
# Nếu sinh viên có đrl ở những dòng khác thì điền trung bình của những dòng đó
# Nếu sinh viên không có dòng nào có đrl thì điền trung bình của cột
col_mean = merge_df4["drl"].mean()
merge_df4["drl"] = merge_df4.groupby("mssv")["drl"].transform(lambda x: x.fillna(x.mean() if x.notna().any() else col_mean))
merge_df4.head()
```

```python
# Tạo cột số thứ tự của học kỳ
merge_df4["hocky_sx"] = merge_df4.groupby("mssv").cumcount()+1
merge_df4.head()
```

# Bảng chính

```python
# dataframe chính
df = merge_df4.copy()
```

```python
# Điểm trung bình của các học kì trước
sum_dtbtl = df.groupby("mssv")["dtbhk"].cumsum().shift()
sum_sohk = df.groupby('mssv')['hocky_sx'].cumcount()
df["dtbtl"] = sum_dtbtl/sum_sohk
df["dtbtl"].head()
```

```python
# Điểm rèn luyện trung bình của các học kì trước
sum_drltl = df.groupby("mssv")["drl"].cumsum().shift()
sum_sohk = df.groupby('mssv')['hocky_sx'].cumcount()
df["drltl"] = sum_drltl/sum_sohk
df["drltl"].head()
```

```python
# Điểm trung bình học kì trước
df['dtbhk_truoc'] = df.groupby('mssv')['dtbhk'].shift(1)
df["dtbhk_truoc"].head()
```

```python
# Điểm rèn luyện học kì trước
df['drlhk_truoc'] = df.groupby('mssv')['drl'].shift(1)
df["drlhk_truoc"].head()
```

```python
# Tổng tín chỉ đã học tính đến học kì trước
df['sotctl'] = df.groupby('mssv')['sotchk'].cumsum().shift()
df["sotctl"].head()
```

# Tính số môn không đạt của kì trước

```python
df_diemhp = pd.read_csv("../data_clean/diem_Thu.csv")
df_diemhp.head()
```

```python
# Lọc những dòng có mssv, năm học, học kỳ tồn tại trong bảng điểm trung bình học kỳ
filtered_df = pd.merge(df, df_diemhp, on=["mssv", "namhoc", "hocky"], how="left")[[col for col in df_diemhp.columns]+["hocky_sx"]]
filtered_df.shape
```

```python
# Điền trống những giá trị diem_hp
filtered_df["diem_hp"].fillna(0, inplace=True)
```

```python
# Tạo cột số môn không đạt
count = filtered_df.groupby(['mssv', 'namhoc', 'hocky']).apply(lambda x: ((x['diem_hp'] < 5) & (~x['trangthai'].isin([0,4,5]))).sum())

# reset index -> chuyển về dataframe thông thường
count_df = count.reset_index(name='somon_khongdat')
```

```python
count_df.shape
```

```python
# Thêm cột số môn không đạt của học kì trước
count_df['somon_khongdat_hktruoc'] = count_df.groupby('mssv')['somon_khongdat'].shift(1)
count_df.head(20)
```

# Final data

```python
# Gộp bảng df (đrb, đrl) với bảng tính số môn không đạt
main = pd.merge(df, count_df, on=["mssv", "namhoc", "hocky"], how="left")
main.shape
```

```python
# Tạo danh sách các khoảng giá trị và xếp loại tương ứng

# {Kém, Yếu}, Trung bình, Khá, {Giỏi, Xuất sắc}
bins = [0, 5, 7, 8, 10.1]
labels = [0, 1, 2, 3]

# Áp dụng pd.cut để xếp loại giá trị của cột "diemtb"
main["xeploai"] = pd.cut(main["dtbhk"], bins=bins, labels=labels, include_lowest=True, right=False)
main.drop(columns="dtbhk", inplace=True)
```

```python
def convert_DRL_to_xeploai(x):
  if x < 35:
    return 0
  elif x >= 35 and x < 50:
    return 1
  elif x >= 50 and x < 65:
    return 2
  elif x >= 65 and x < 80:
    return 3
  elif x >= 80 and x < 90:
    return 4
  elif x >= 90:
    return 5
```

```python
main["drltl"] = main["drltl"].apply(lambda x: convert_DRL_to_xeploai(x))
main["drlhk_truoc"] = main["drlhk_truoc"].apply(lambda x: convert_DRL_to_xeploai(x))
```

```python
main["xeploai"] = main["xeploai"].astype("int64")
```

```python
# One hot giá trị khoa
one_hot_df = pd.get_dummies(main['khoa'], prefix='khoa')

main = main.iloc[:, :2].join(one_hot_df.astype(int)).join(main.iloc[:, 2:])
main.drop(columns=["khoa"], inplace=True)
```

```python
# One hot giá trị hệ đào tạo
one_hot_df = pd.get_dummies(main['hedt'], prefix='hedt')

main = main.iloc[:, :2].join(one_hot_df.astype(int)).join(main.iloc[:, 2:])
main.drop(columns=["hedt"], inplace=True)
```

```python
# Bỏ cột drl
main.drop(columns=["drl"], inplace=True)
```

```python
# Bỏ cột số môn không đạt hk hiện tại
main.drop(columns=["somon_khongdat"], inplace=True)
```

```python
# Bỏ các dòng học kỳ 1
main.drop(main[main["hocky_sx"]==1].index, inplace=True)
main.shape
```
